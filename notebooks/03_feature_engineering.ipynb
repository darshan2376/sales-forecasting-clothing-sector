```python
# 03_feature_engineering.ipynb
import pandas as pd

df = pd.read_csv('../data/cleaned_sales_data.csv', parse_dates=['date'])
df.sort_values('date', inplace=True)
df.set_index('date', inplace=True)

# Add moving averages
df['revenue_rolling_7'] = df['revenue'].rolling(window=7).mean()
df['revenue_rolling_30'] = df['revenue'].rolling(window=30).mean()

df.reset_index().to_csv('../data/cleaned_sales_data.csv', index=False)
df.tail()
```
