```python
# 01_data_cleaning.ipynb
import pandas as pd

# Load raw data
df = pd.read_csv('C:/Sales-Forecasting-Clothing-Sector/data/raw_sales_data.csv', parse_dates=['Date'])

# Rename for consistency
df.columns = ['date', 'store', 'category', 'units_sold', 'revenue']

# Basic cleaning
df.dropna(inplace=True)
df['units_sold'] = df['units_sold'].astype(int)
df['revenue'] = df['revenue'].astype(float)

# Save cleaned data
df.to_csv('../data/cleaned_sales_data.csv', index=False)

# Preview
df.head()
```
